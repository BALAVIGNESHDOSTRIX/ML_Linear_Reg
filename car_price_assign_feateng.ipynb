```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)

df = pd.read_csv('car_price_assign.csv', delimiter=',')
df.head()
```

We will be performing all the below steps in Feature Engineering

  1.Missing values
  2.Temporal variables
  3.Categorical variables: remove rare labels
  4.Standarise the values of the variables to the same range

```python
# Capture if any missing values in the features

df.isnull().sum()
```

Show the dataset don't having any null values

```python
# Handling the Rare Categorical Variables
categ_features = df.columns[df.dtypes == 'O']
categ_features
```

```python
for x in categ_features:
  temp=df.groupby(x)['price'].count()/len(df)
  tempdf = temp[temp>0.01].index
  df[x] = np.where(df[x].isin(tempdf), df[x], 'Rar_var')
```

```python
df[categ_features]
```

```python
for x in categ_features:
  print(x,':',len(df[x].unique()), 'labels')
```

```python
# So No need to select top 10 categories

categ_df = pd.get_dummies(df[categ_features])
```

```python
categ_df.shape
```

```python
categ_df.head()
```

```python
df = df.drop(categ_features, axis=1)
```

```python
df.drop('car_ID', axis=1, inplace=True)
```

```python
df.shape
```

```python
newdf = pd.concat([df, categ_df], axis=1)
```

```python
newdf.head()
```

```python
# Feature Scaling

feaure_scale = [x for x in newdf.columns if x != 'price']
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
scalar.fit(newdf[feaure_scale])
```

```python
data = pd.concat([newdf['price'].reset_index(drop=True),
                    pd.DataFrame(scalar.transform(newdf[feaure_scale]), columns=feaure_scale)],
                    axis=1)
```

```python
data.head()
```

```python
data.to_csv('car_price_assign_processed.csv',index=False)
```
